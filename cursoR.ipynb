```r
#Analisis de numero de asesinatos por ciudad en los Estados Unidos.
#Carga de librerias necesarias
library(tidyverse)
  library(dslabs)
  library(dplyr)
```

```r
#Carga de data de asesinatos, necesaria para el analisis 
data(murders)
```

```r
#Visualizar tabla 
murders
```

```r
#Grafica de las regiones con mas asesinatos en los estados unidos, asesinatos vs region (Norte, sur, este y oeste)  
murders %>%
    ggplot(aes(population, total, label = abb, color = region)) +
    geom_label()
```

```r
#Grafica de las regiones con mas asesinatos en los estados unidos, asesinatos vs region (Norte, sur, este y oeste)  
murders %>%
    ggplot(aes(region, total, label = abb, color = region)) +
    geom_label()
```

```r
#Grafica relación entre la cantidad de habitantes y numero de asesinatos en la ciudad.  
murders %>%
    ggplot(aes(total, population, label = abb)) +
    geom_label()

#Podemos observar que LA, siendo una ciudad con pocos habitantes en comparacion al resto, presenta el mayor % de asesinatos en proporcion a sus habitantes. 
```

```r
murders <- mutate(murders, rate=total/population*100000)
murders
```

```r

murders %>% 
    ggplot(aes(population, rate, label = abb, color = region)) +
    geom_label()
```
