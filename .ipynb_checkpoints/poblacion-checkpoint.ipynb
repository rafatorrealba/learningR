```r
library(tidyverse)
library(dslabs)
library(nycflights13, gapminder)
library(dplyr)
```

```r
lapintana <- read_csv("data/lapintana.csv")
```

```r
lapintana
```

```r
ggplot(data = lapintana) +
  geom_label(mapping = aes(x = Fecha, y = Hombres, label = Hombres))
```

```r
ggplot(data = lapintana) +
  geom_line(mapping = aes(x = Fecha, y = Hombres))
```

```r
ggplot(data = lapintana) +
  geom_path(mapping = aes(x = Fecha, y = Hombres,))
```
